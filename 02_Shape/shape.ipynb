```python
import math
import numpy as np
import matplotlib.pyplot as plt # low level graph plotting library in python

class Shape:
    def __init__(self, name = "shape"):
        self._name = name
        
    def area():
        raise NotImplementedError ("Each shape must implement its own area.") #overidden

    def plot():
        raise NotImplementedError ("Each shape must implement its own plot")

    def get_name(self):
        return self._name

#CIRCLE CLASS:

class Circle(Shape):
    def __init__(self, radius):
        super().__init__("Circle")
        self.radius = radius

    def area(self):
        return math.pi*self.radius**2

    def plot(self):
        theta = np.linspace(0, 2*math.pi,100)
        x = self.radius * np.cos(theta)
        y = self.radius * np.sin(theta)
        plt.plot(x,y)
        plt.title("Circle")
        plt.axis("Equal")
        plt.show()

#TRIANGLE CLASS : EQUILATERAL, ISOSCELES, RIGHT, SCALENE

class Triangle(Shape):
    def __init__(self, a, b, c):
        super().__init__("Triangle")
        self.a = a
        self.b = b
        self.c = c

    def area(self):
        #Heron's Formula:
        s = (self.a + self.b + self.c) / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))

    def plot(self):
        #Triangle plot  
        x = [0, self.a, self.b / 2, 0]
        y = [0, 0, self.area() * 2/self.a, 0]
        plt.plot(x, y)
        plt.title("Triangle")
        plt.axis("Equal")
        plt.show()

#HEXAGON CLASS 

class Hexagon(Shape):
    def __init__(self,side):
        super().__init__("Hexagon")
        self.side = side

    def area(self):
        return (3 * math.sqrt(3) / 2)* self.side ** 2

    def plot(self):
        theta = np.linspace(0, 2*math.pi, 7)
        x = self.side * np.cos(theta)
        y = self.side * np.sin(theta)
        plt.plot(x, y)
        plt.title("Hexagon")
        plt.axis("Equal")
        plt.show

#CONE CLASS

class Cone(Shape):
    def __init__(self, radius, height):
        super().__init__("Cone")
        self.radius = radius
        self.height = height

    def area(self):
        l = math.sqrt(self.radius**2 + self.height**2)
        return math.pi * self.radius * (self.radius - 1) #Total Surface Area

    def plot(self):
        x = [-self.radius, self.radius, 0, -self.radius]
        y = [0, 0, self.height, 0]
        plt.plot(x,y)
        plt.title("Cone : (2D VIEW)")
        plt.axis("Equal")
        plt.show()

    #POLYNOMIAL CLASS : TO FIND AREA UNDER THE CURVE

class Polynomial:
    def __init__(self, coeffs):
        self.coeffs = coeffs

    def __call__(self, x):
        result = 0 
        for power, coeff in enumerate(self.coeffs):
            result += coeff * (x ** power)
        return result

    def area(self, a, b, n = 1000):
        x = np.linspace(a, b, n)
        y = self.__call__(x)
        area = np.trapz(y, x)
        return area

    def plot(self, a, b):
        x = np.linspace(a, b, n)
        y = self.__call__(x)
        plt.plot(x, y)
        plt.title("Polynomial Curve")
        plt.show()

if __name__ == "__main__":

    #Circle
    c = Circle(5)
    print("Circle Area:", c.area())
    c.plot()

    #Triangle
    t = Triangle(3, 4,5)
    print("Triangle Area:", t.area())
    t.plot()

    #Hexagon 
    h = Hexagon(4)
    print("Hexagon Area:",h.area())
    h.plot()

    #Cone 
    cone = Cone(3, 6)
    print("Cone Surface Area:", cone.area())
    cone.plot()

    #Polynomial: 2 + 3x + x^2
    p = Polynomial([2, 3, 1])
    print("Area under polynomial:", p.area(0, 5))
    p.plot(0, 5)

    
    
        

        
```
